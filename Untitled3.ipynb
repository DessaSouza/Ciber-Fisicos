```python
import numpy as np
import matplotlib.pyplot as plt
import random as rd


def plot_trig_circle(angle):
    # Converte o ângulo para radianos
    angle_rad = np.deg2rad(angle)

    # Define o círculo trigonométrico
    circle = plt.Circle((0, 0), 1, color='black', fill=False)

    # Cria a figura e os eixos
    fig, ax = plt.subplots(figsize=(8, 8))

    # Adiciona o círculo aos eixos
    ax.add_artist(circle)

    # Define os limites dos eixos
    ax.set_xlim(-1.5, 1.5)
    ax.set_ylim(-1.5, 1.5)

    # Adiciona os eixos cartesianos
    ax.axhline(0, color='black', linewidth=0.5)
    ax.axvline(0, color='black', linewidth=0.5)

    # Plota o ângulo escolhido
    ax.plot([0, np.cos(angle_rad)], [0, np.sin(angle_rad)], color='red', label='Ângulo = {}°'.format(angle))


    if np.cos(angle_rad) != 0:
        ax.plot([1, 1], [0, np.tan(angle_rad)], color='magenta', label=f'Tangente = {np.tan(angle_rad):.3f}') #Tangente
        ax.plot([np.cos(angle_rad), 1], [np.sin(angle_rad), np.tan(angle_rad)], color='black', linestyle='--')  # Linha pontilhada
    ax.plot([np.cos(angle_rad), np.cos(angle_rad)], [0, np.sin(angle_rad)], color='blue',label=f'Seno = {np.sin(angle_rad):.3f}')  # Seno
    ax.plot([0, np.cos(angle_rad)], [0, 0], color='green', label=f'Cosseno = {np.cos(angle_rad):.3f}') #Cosseno

    # Adiciona legenda
    ax.legend()

    # Mostra o gráfico
    plt.show()


# Solicita ao usuário o ângulo desejado

angle = float(input("Por favor, insira o ângulo desejado em graus (0 < θ < 180): "))

# Verifica se o ângulo está no intervalo válido
if 0 < angle < 180:
    plot_trig_circle(angle)
else:
    print("Ângulo inválido. Por favor, insira um ângulo no intervalo (0 < θ < 180).")
```
